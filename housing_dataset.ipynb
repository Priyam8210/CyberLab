```python
def rel_feat(dataset, threshold):
  col_cor = set()
  cor_mat = dataset.corr()
  for i in range(len(cor_mat.columns)):
    for j in range(i):
      if(abs(cor_mat.iloc[i,j])>threshold):
        if (abs(cor_mat.iloc[i]["SalePrice"])>abs(cor_mat.iloc[i,j])):
          col_cor.add(cor_mat.columns[j])
        else:
          col_cor.add(cor_mat.columns[i])
  return col_cor
```

# SVM
  SVM stands for Support Vector Machine. It's a supervised machine learning algorithm which work to find best hyperplane(decision boundry) among different classes. SVM performs on non-linear classificatin using **KERNAL TRICK**, which maps their input into high dimensional feature space.

### 3 Hyperparameters
**1. C**

>   C is a hyperparameter in SVM to control error. If the value of C is high, means high error and if the value of C is low means low error








   

**2. Gamma**


> Gamma is hyperparameter which decides  the curvature in decision boundary.  If we have high value of Gamma means more curvature and vice-versa. Gamma is used when we use the Gaussian RBF Kernal. If we use linear or polynomial kernal, then only hyperparameter C is enough.



    

**3. Kernals**


> Working on non-linear dataset, 'Kernal Trick' makes it easier to classifies the points. Different types of kernals are Polynomial kernal, sigmoid kernal, RBF kernal, Bessel function kernal, Anova kernal, etc.



## Advantage
-> SVM works better when data is linear.

-> It is more effective on high dimension.

-> With help of kernal trick, we can solve complex data also.

-> It is not sensetive to outliers


## Disadvantage
-> For perfect decision boundary, choosing good kernal is not easy.

-> It doesn't works well on big dataset.

-> Hyperparameter tuning is not so easy, especially in case of Cost-C and gamma. It hard to visualize their impact.
