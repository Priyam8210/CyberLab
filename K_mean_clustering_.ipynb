```python
def ecludian(X, cent):
  return np.sqrt(np.sum(((X-cent)**2)))
```

```python
def close_centroid(s, cen):
  dist = [ecludian(s, pt) for pt in cen]
  close_id = np.argmin(dist)
  return close_id
```

```python
def init_cluster(X, centroid, K):
  cluster = [[] for _ in range(K)]
  for i, sample in enumerate(X):
    cen_id = close_centroid(sample, centroid)
    cluster[cen_id].append(i)
  return cluster
```

```python
def centroid_form(X, cluster, K):
  n1 = np.size(X, 1)
  centroid = np.zeros((K, n1))
  for i, cluster in enumerate(cluster):
    mean = np.mean(X[cluster], axis = 0)
    centroid[i] = mean
    return centroid
```

```python
def get_label(cluster, n):
  label = np.empty(n)
  for id_c, cl in enumerate(cluster):
    for sa_id in cl:
      label[sa_id] = id_c
  return label
```
